#### Going from 133 features to 55 features by removing ones that aren't contributing

```python
import pandas as pd
```

```python
df = pd.read_csv('Dataset/train_selected.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
for i in range(133):
    print("Column: ", i, "\n", df.iloc[:,i].value_counts(), "\n")
# Shows that a lot of columns are unnecessary. Time for RFE then!
```

```python
# Before that, lets remove columns that have a single value
uniqueValues = df.nunique()
```

```python
# Making a list of non singluar value columns
nonSingular = []
for i in range(len(uniqueValues)):
    if uniqueValues[i] != 1:
        nonSingular.append(df.columns[i])
```

```python
len(nonSingular)
```

```python
df = df[nonSingular]
```

```python
df.shape
```

```python
df.head()
```

```python
df.to_csv('./Dataset/train_selected_cleaned.csv', index=False)
# So we have removed the columns that were not contributing any information, lets apply 
# to this and see how it performs, we'll reduce features after that
```
